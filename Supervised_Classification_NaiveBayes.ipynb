```python
import pandas as pd 
link = '../../DATASET/dataset.csv'
DF_FOOTBALL = pd.read_csv(link)
```

```python
DF_FOOTBALL.head()
```

```python
DF_FOOTBALL.info()
```

```python
kolom_penting = ['Age','Height','Handling ']
```

```python
kolom_tidak_penting = []
for col in DF_FOOTBALL.columns :
    if col != 'Age' and col != 'Height' and col != 'Handling' :
        kolom_tidak_penting.append(col)
```

```python
for col in DF_FOOTBALL.columns :
    if col != 'Age' and col != 'Height' and col != 'Handling' :
        DF_FOOTBALL.drop([col],axis=1,inplace=True)
```

```python
DF_FOOTBALL.columns
```

```python
DF_FOOTBALL
```

```python
filtered_data = DF_FOOTBALL[(DF_FOOTBALL['Age']<=30) & (DF_FOOTBALL['Height']>=175) & (DF_FOOTBALL['Handling']>=5)]
filtered_data['Target'] = 1
unfiltered_data = DF_FOOTBALL[(DF_FOOTBALL['Age']>30) | (DF_FOOTBALL['Height']<175) | (DF_FOOTBALL['Handling']<5)]
unfiltered_data['Target'] = 0
```

```python
filtered_data
```

```python
unfiltered_data
```

```python
UNION_DF = pd.concat([filtered_data, unfiltered_data])
```

```python
UNION_DF
```

```python
DF_FEATURE = UNION_DF.drop(['Target'], axis=1)
DF_TARGET = UNION_DF['Target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(DF_FEATURE,DF_TARGET, test_size=0.2, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train,y_train)
```

```python
ypred = model.predict(X_test)
ypred
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, ypred))
```

```python
from sklearn.model_selection import cross_val_score,KFold
cv = KFold(n_splits=10, random_state=1, shuffle=True)
scores = cross_val_score(model, DF_FEATURE, DF_TARGET, scoring='accuracy', cv=cv, n_jobs=-1)
print(scores)
```
