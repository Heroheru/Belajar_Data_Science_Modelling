<h1 style="text-align:center">SUPPORT VECTOR MACHINE</h1>

<p align="justify">Support Vector Machine (SVM) merupakan salah satu metode dalam supervised learning yang biasanya digunakan untuk klasifikasi (seperti Support Vector Classification) dan regresi (Support Vector Regression). Dalam pemodelan klasifikasi, SVM memiliki konsep yang lebih matang dan lebih jelas secara matematis dibandingkan dengan teknik-teknik klasifikasi lainnya. SVM juga dapat mengatasi masalah klasifikasi dan regresi dengan linear maupun non linear.

<b>SVM digunakan untuk mencari hyperplane terbaik dengan memaksimalkan jarak antar kelas.</b> Hyperplane adalah sebuah fungsi yang dapat digunakan untuk pemisah antar kelas. Dalam 2-D fungsi yang digunakan untuk klasifikasi antar kelas disebut sebagai line whereas, fungsi yang digunakan untuk klasifikasi antas kelas dalam 3-D disebut plane similarly, sedangan fungsi yang digunakan untuk klasifikasi di dalam ruang kelas dimensi yang lebih tinggi di sebut hyperplane.</p>

<h2>7). Mengubah data NON NUMERIC menjadi data NUMERIC dengan encoding karena machine learning harus numeric datanya supaya dapat diproses</h2>

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
for col in DF_JAMUR_feature :
    print(col, DF_JAMUR_feature[col].unique())
```

<h2>8). Agar dapat memprediksi data baru, maka data baru tersebut harus di encode juga berdasarkan hasil encoding pada data lama sehingga pasangan hasil encoding pada data lama harus disimpan dalam bentuk dictionary agar encoding pada data baru dapat dilakukan</h2>

```python
for col in DF_JAMUR_feature.columns:
    temp_keys = DF_JAMUR_feature[col].values
    temp_values = label_encoder.fit_transform(DF_JAMUR_feature[col])
    dict_temp = dict(zip(temp_keys, temp_values))
    dict_all[col] = dict_temp
```

```python
for col in DF_JAMUR_feature.columns :
    print(col,dict_all[col])
```

```python
for col in DF_JAMUR_feature.columns:
    DF_JAMUR_feature.replace(dict_all[col], inplace=True)
 
```

<h2>9). Membagi dataset 80:20</h2>

```python
DF_JAMUR_feature
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(DF_JAMUR_feature, DF_JAMUR_label, test_size = 0.2, random_state = 0)
```

<h2>10). Melatih dan Menguji Model</h2>

```python
from sklearn import svm
```

```python
#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

<h2>11). Mengevaluasi Model</h2>

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
print("F1score:",metrics.f1_score(y_test, y_pred))
print("Precission:",metrics.precision_score(y_test, y_pred))
```
